```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
data
```

```python
data.drop(data[data['is_wicket']<1].index, inplace = True)
```

```python
bowler = ['SL Malinga','A Mishra']
decision_making = pd.DataFrame([], columns=['Bowler','Wicket Type','Times'])
for id,element in enumerate(bowler):
    cought = data[(data['bowler']==element) & (data['dismissal_kind'] == 'caught')]
    run_out = data[(data['bowler']==element) & (data['dismissal_kind'] == 'run out')]
    bowled = data[(data['bowler']==element) & (data['dismissal_kind'] == 'bowled')]
    
    lbw = data[(data['bowler']==element) & (data['dismissal_kind'] == 'lbw')]
    stumped = data[(data['bowler']==element) & (data['dismissal_kind'] == 'stumped')]
    caught_and_bowled = data[(data['bowler']==element) & (data['dismissal_kind'] == 'caught and bowled')]
    hit_wicket = data[(data['bowler']==element) & (data['dismissal_kind'] == 'hit wicket')]
       
# append to decision making
    decision_making = decision_making.append({'Bowler':element,'Wicket Type':'cought', 
                                              'Times':cought['bowler'].count()}, ignore_index=True)
    decision_making = decision_making.append({'Bowler':element,'Wicket Type':'run out', 
                                              'Times':run_out['bowler'].count()}, ignore_index=True)
    decision_making = decision_making.append({'Bowler':element,'Wicket Type':'bowled', 
                                              'Times':bowled['bowler'].count()}, ignore_index=True)
    
    decision_making = decision_making.append({'Bowler':element,'Wicket Type':'lbw', 
                                              'Times':lbw['bowler'].count()}, ignore_index=True)
    decision_making = decision_making.append({'Bowler':element,'Wicket Type':'stumped', 
                                              'Times':stumped['bowler'].count()}, ignore_index=True)
    decision_making = decision_making.append({'Bowler':element,'Wicket Type':'caught and bowled', 
                                              'Times':caught_and_bowled['bowler'].count()}, ignore_index=True)
    decision_making = decision_making.append({'Bowler':element,'Wicket Type':'hit wicket', 
                                              'Times':hit_wicket['bowler'].count()}, ignore_index=True)
    
   
```

```python
decision_making.drop(decision_making[decision_making['Times']<1].index, inplace = True)
```

```python
decision_making
```

```python
sns.catplot(x="Wicket Type", y="Times", hue="Bowler", data = decision_making, kind='bar',height=5,aspect=2)
plt.title("Bolwlers compare")
plt.xlabel("Wicket Type")
plt.ylabel("Times")
plt.show()
```
